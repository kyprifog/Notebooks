```python
import json
import requests
import pandas as pd
import numpy as np
import math
import pandas as pd
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
import matplotlib
matplotlib.use('TkAgg')
```

```python
response = requests.get("https://covidtracking.com/api/us/daily")
data = response.json()
df = pd.DataFrame(data=data)
df['date']  = pd.to_datetime(df['date'], format='%Y%m%d')
```

```python
def plot(starting_pop, xlim, ylim, beta):
    # Total population, N.  US population
    N = 327200000
    # Initial number of infected and recovered individuals, I0 and R0.
    #  Start at day 13 with 5000 infected
    I0, R0 = starting_pop, 2.2
    # Everyone else, S0, is susceptible to infection initially.
    S0 = N - I0 - R0
    # Contact rate, beta, and mean recovery rate, gamma, (in 1/days).
    beta, gamma = beta, 1./8
    # A grid of time points (in days)
    t = np.linspace(0, 1000, 1000)

    # The SIR model differential equations.
    def deriv(y, t, N, beta, gamma):
        S, I, R = y
        dSdt = -beta * S * I / N
        dIdt = beta * S * I / N - gamma * I
        dRdt = gamma * I
        return dSdt, dIdt, dRdt

    # Initial conditions vector
    y0 = S0, I0, R0
    # Integrate the SIR equations over the time grid, t.
    ret = odeint(deriv, y0, t, args=(N, beta, gamma))
    S, I, R = ret.T

    # Plot the data on three separate curves for S(t), I(t) and R(t)
    fig = plt.figure(facecolor='w')
    ax = fig.add_subplot(111, axisbelow=True)
    ax.plot(t, S, 'b', alpha=0.5, lw=2, label='Susceptible')
    ax.plot(t, I, 'r', alpha=0.5, lw=2, label='Infected')
    ax.plot(t, R, 'g', alpha=0.5, lw=2, label='Recovered with immunity')
    ax.plot(t, I*0.034, "black", alpha=1, lw=2, label='Death Rate')
    ax.set_xlabel('Time /days')
    ax.set_ylabel('Number')

    # in 4 days went to 20,000 infected
    ax.set_ylim(0,ylim)
    ax.set_xlim(0,xlim)
    ax.yaxis.set_tick_params(length=0)
    ax.xaxis.set_tick_params(length=0)
    ax.grid(b=True, which='major', c='w', lw=2, ls='-')
    legend = ax.legend()
    legend.get_frame().set_alpha(0.5)
    for spine in ('top', 'right', 'bottom', 'left'):
        ax.spines[spine].set_visible(False)
    fig.show()
```

```python
# on 14th day had starting infected of 5000 and in 4 days had 20000
plot(5000,4,20000, 0.5)
df.plot(x ='date', y='positive')
```
